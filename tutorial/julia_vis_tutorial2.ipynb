```julia
#frequency hsitogram
using Plots

x = randn(10^3)
histogram(x)
```

```julia
using RDatasets
using Plots
cars = dataset("datasets", "mtcars")
bar(cars.Model,
    # plotting we want to plot attributes
    [cars.MPG,cars.QSec], 
    label = ["Miles/Gallon" "QSec"],
    title = "Models-Miles/Gallon and Qsec",
    #x axis tick label angle
    xrotation = 90,
    #size of the figure 
    size = [600, 400],
    legend =:topleft,
    xticks =:all,
    #number of ticks shown
    yticks =0:35)
```

Weighted histogram

```julia
f_exp(x) = exp(x)/(exp(1)-1)

x = rand(10^4)
w = exp.(x)

histogram(x, label="Experimental", bins=:scott, weights=w, normalize=:pdf, color=:gray)
#plot!(f_exp, label="Analytical", lw=3, color=:red)
#plot!(legend=:topleft)
xlims!(0, 1.0)
ylims!(0, 1.6)
title!("Uniform distribution, weighted by exp(x)")
xlabel!("x")
ylabel!("P(x)")
```

```julia
x = rand(10^4)
histogram(x, label="Experimental", bins=:30,color=:gray)
#plot!(f_exp, label="Analytical", lw=3, color=:red)
#plot!(legend=:topleft)
xlims!(0, 1.0)
ylims!(0, 1.6)
title!("Uniform distribution, weighted by exp(x)")
xlabel!("x")
ylabel!("P(x)")
```

```julia
#2D Heatmap 
x = randn(10^4)
y = randn(10^4)
histogram2d(x, y)
```

```julia
#normalize the values and show empty bins 
histogram2d(x, y, bins=(20, 20), show_empty_bins=true,
    normalize=:pdf, color=:plasma)
title!("Normalized 2D Histogram")
xlabel!("x")
ylabel!("y")
```
